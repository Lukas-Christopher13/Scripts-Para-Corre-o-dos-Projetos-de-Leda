Criar um novo campo de data completa chamado full_date no formato DD/MM/AAAA de acordo com os dados das colunas date e year.

```python
import numpy as np
import pandas as pd

mathces_T1_path = "Ligas_Europeias_de_Futebol/transformacoes/matches_T1.csv"
mathces_T2_path = "Ligas_Europeias_de_Futebol/transformacoes/matches_T2.csv"

df_read = pd.read_csv(mathces_T1_path, low_memory=False)

dates = df_read["date"]
years = df_read["year"]

meses_para_numeros = {
    'January' : '01',
    'February' : '02',
    'March' : '03',
    'April' : '04',
    'May' : '05',
    'June' : '06',
    'July' : '07',
    'August' : '08',
    'September' : '09',
    'October' : '10',
    'November' : '11',
    'December' : '12'
}

my_full_date = []
#Gera minhas datas
for date, year in zip(dates, years) :
    splited_date = date.split(" ")
    if len(splited_date[2]) == 1:
        splited_date[2] = '0' + splited_date[2]
    formated_date = f"{splited_date[2]}/{meses_para_numeros[splited_date[1]]}/{year}"

    my_full_date.append(formated_date)

mathces_T2 = pd.read_csv(mathces_T2_path, low_memory=False) 
full_date = mathces_T2["full_date"]

#Compara as minhas datas com as datas geradas pelos alunos
cont = 2
for date, my_date in zip(full_date, my_full_date):
    if(date != my_date):
        print(f"{cont}- As datas não batem! {date} | {my_date}")
    cont+=1
print("Os dados batem!")
```

Filtrar apenas os jogos da liga inglesa "English Premier League" através do campo league.

```python
import numpy as np
import pandas as pd

mathces_T2_path = "Ligas_Europeias_de_Futebol/transformacoes/matches_T2.csv"
mathces_F1_path = "Ligas_Europeias_de_Futebol/transformacoes/matches_F1.csv"

df_read = pd.read_csv(mathces_T2_path, low_memory=False)

resultado_filtrado = df_read[df_read["league"].str.contains("English Premier League")]

resultado_dos_alunos = pd.read_csv(mathces_F1_path, low_memory=False)

print(f"Tamaho do arquivo que deve ser gerado: {len(resultado_filtrado)}")
print(f"Tamaho do arquivo gerado: {len(resultado_dos_alunos)}")

#Comparado os ids
meus_resultados_id = resultado_filtrado["id"]
resultado_dos_alunos_id = resultado_dos_alunos["id"]

for meu_id, alunos_id in zip(meus_resultados_id, resultado_dos_alunos_id):
    if(meu_id != alunos_id):
        print(f"Erro em -> Meu id:{meu_id} | Aluno id: {alunos_id}")
print("Os arquivos são iguais!")

```

Filtrar os jogos que tiveram attendance superior a 20.000 pagantes.

```python
import numpy as np
import pandas as pd

mathces_F2_path = "Ligas_Europeias_de_Futebol/transformacoes/matches_F2.csv"

df_read = pd.read_csv(mathces_F2_path, low_memory=False)

resultados = df_read["attendance"]

for resultado in resultados:
    if(int(resultado.replace(",", "")) < 20000):
        print(int(resultado.replace(",", "")))

print("Todos os valores são maiores que 20000!")
```
